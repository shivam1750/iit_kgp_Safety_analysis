```python
gold_data['Date'] = pd.to_datetime(gold_data['Date'])
silver_data['Date'] = pd.to_datetime(silver_data['Date'])
```

```python
gold_data.set_index('Date', inplace=True)
silver_data.set_index('Date', inplace=True)
```

```python
gold_monthly = gold_data.resample('M').mean()
silver_monthly = silver_data.resample('M').mean()
```

```python
plt.figure(figsize=(25,12))
plt.plot(gold_monthly.index, gold_monthly['Price'], label='Gold')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.title('Monthly Gold Prices')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(25,12))
plt.plot(silver_monthly.index, silver_monthly['Price'], label='Silver')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.title('Monthly Silver Prices')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(25,12))
plt.plot(gold_monthly.index, gold_monthly['Price'], label='Gold')
plt.plot(silver_monthly.index, silver_monthly['Price'], label='Silver')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.title('Monthly Gold and Silver Prices')
plt.legend()
plt.show()
```
