Predict the months of 2023 where export of the minerals: gold  will yield maximum profits in a visual format.



```python
plt.figure(figsize=(25,12))
plt.plot(gold_data.index,gold_data['Price'], label='Gold')

plt.title('Monthly Gold Profits for Mineral Exports in 2023')
plt.legend()
plt.xlabel('Month')
plt.ylabel('Profits (USD)')
plt.show()
```
